```python
import pandas as pd
```

```python
##Importa o documento base
df = pd.read_csv('PYTHON_DIO/olist_customers_dataset.csv')
df
```

```python
##renomeia colunas
df.rename(columns={"customer_zip_code_prefix": "cep_prefixo", "customer_city": "comprador_cidade", "customer_state": "comprador_estado"}, inplace=True)
df
```

```python
##mostra o nome das colunas
df.columns
```

```python
##mostra o tipo de dados nas colunas
df.dtypes
```

```python
##apaga colunas
df2 = df.drop(columns="customer_unique_id")
df2
```

```python
##agrupa e conta o numero de compradores por estado
df2.groupby("comprador_estado")["customer_id"].count()
```

```python
##Adiciona outra fonte de dados
df3 = pd.read_csv('PYTHON_DIO/olist_orders_dataset.csv')
df3
```

```python
df3.dtypes
```

```python
##agrupa e conta por status unico de vendas
df3.groupby("order_status")["order_id"].nunique()
```

```python
##verficia valores nulos
df2.isnull().sum()
```

```python
df3.isnull().sum()
```

```python
df4 = df3.drop(columns=["order_approved_at", "order_delivered_carrier_date", "order_estimated_delivery_date"])
df4
```

```python
df4.isnull().sum()
```

```python
##Apaga as linhas com valores nulos
##Antes eram 99441 rows agora são 96476 rows
df4.dropna(inplace=True)
df4
```

```python
import datetime as dt
```

```python
##Passando a coluna de data para o tipo data
df4["order_purchase_timestamp"] = pd.to_datetime(df4["order_purchase_timestamp"])
```

```python
df4.dtypes
```

```python
##Adiciona a coluna de ano no dataframe
df4["purchase_year"] = df4["order_purchase_timestamp"].dt.year
df4
```

```python
##agrupa a soma de vendas por ano
df4.groupby(df4["purchase_year"]).sum()
```

```python
##Adiciona a coluna de mes e coluna de dia no dataframe
df4["purchase_month"], df4["purchase_day"] = df4["order_purchase_timestamp"].dt.month, df4["order_purchase_timestamp"].dt.day
df4
```

```python
##Vendas de janeiro de 2018
vendas_janeiro_2018 = df4.loc[(df4["purchase_year"] == 2018) & (df4["purchase_month"] == 1)]
vendas_janeiro_2018
```
